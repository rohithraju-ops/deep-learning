```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([2,3])
y = np.array([1,-1])

print(x+y)
print(x-y)
print(2*x)

print(np.dot(x,y))

# Positive = same direction, Negative = opposite, Zero = perpendicular
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='x')
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='red', label='y')
ax.set_xlim(-2, 4)
ax.set_ylim(-3, 4)
ax.grid(True)
ax.legend()
ax.set_title("Two Vectors")
plt.show()
```

```python
# Matrix of 2D array

# In neural nets: W is a matrix that TRANSFORMS your input

W = np.array([[1, 2],
              [3, 4],
              [5, 6]])   #  3 neurons, 2 inputs each

X = np.array([1.0, 2.0])  #  2 inputs

# Matrix-vector multiplication
output = W @ X             
print("Output shape:", output.shape)  
print("Output:", output)

# This is IDENTICAL to computing each neuron manually:
neuron1 = W[0] @ X    
neuron2 = W[1] @ X    
neuron3 = W[2] @ X    
print(f"\nManual: {neuron1}, {neuron2}, {neuron3}")
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = np.array([1.0, 2.0])         # shape: (2,)   ← 2 input features
print(f"Input X shape:       {X.shape}")

W1 = np.random.randn(2, 5)       # ← connects 2 inputs to 5 hidden neurons
b1 = np.zeros(5)                 
print(f"W1 shape:            {W1.shape}")

W2 = np.random.randn(5, 1)       # ← connects 5 hidden to 1 output
b2 = np.zeros(1)                
print(f"W2 shape:            {W2.shape}")

# Forward pass — with shape tracking
z1     = X @ W1 + b1            
print(f"\nz1 (pre-activation): {z1.shape}")

hidden = sigmoid(z1)             
print(f"hidden (activated):  {hidden.shape}")

z2     = hidden @ W2 + b2        
print(f"z2 (pre-activation): {z2.shape}")

output = sigmoid(z2)             # shape: (1,) -- single output
print(f"output:              {output.shape}")

# BT DUBS I changed it to 5 to see the shape shift and see what else I need to be shifting
```
